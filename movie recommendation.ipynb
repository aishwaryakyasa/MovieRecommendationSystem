```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
AlluserRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
AlluserRatings.head()
```

```python
corelationMatrix = AlluserRatings.corr()
corelationMatrix.head(5)
```

```python
corelationMatrix = AlluserRatings.corr(method='pearson', min_periods=100)
corelationMatrix.head(5)
```

```python
corelationMatrix['12 Angry Men (1957)'].dropna()
```

```python
particuarUserRatings = AlluserRatings.loc[2].dropna()
particuarUserRatings
```

```python
corelatedMovieValues = pd.Series()

for i in range(0, len(particuarUserRatings.index)):
    print ("Adding sims for " + particuarUserRatings.index[i] + "...")
    
    # Retrieve similar movies to this one that I rated
    # value is noting but entire coloumn named userRatings.index[i]
    # userRatings.index[i] indicates the movie we rated
    value = corelationMatrix[particuarUserRatings.index[i]]
    
    # Now scale its similarity by how well I rated this movie
    # this is just like adding weights according to our review to a particuar movie 
    value = value.map(lambda x: x * particuarUserRatings[i])
    print(value)
    print("........................................................................................")
    # Add the score to the list of similarity candidates
    corelatedMovieValues = corelatedMovieValues.append(value)
    
#Glance at our results so far:
print ("\n******************************************sorting******************************************\n")
corelatedMovieValues.sort_values(inplace = True, ascending = False)
print (corelatedMovieValues)
```

```python
corelatedMovieValues = corelatedMovieValues.groupby(corelatedMovieValues.index).sum()
```

```python
corelatedMovieValues.sort_values(inplace = True, ascending = False)
corelatedMovieValues
```

```python
#inoreder to remove movies which we reviewed from recomendations that are already watched by users
filteredmovies = corelatedMovieValues.drop(particuarUserRatings.index)
filteredmovies
```
